## Exercise 1: Ready made data vs Custom made data

### Q1

**Centola's study (custom-made data):**

Pros:
- Strong internal validity. Because Centola designed the experiment with random assignment to different network structures, he could isolate the effect of network topology on behavior spread. This allows him to make causal claims, the network structure caused differences in how behavior spread.
- Complete data. Since the experiment was purpose-built for the research question, the data contains exactly the variables needed. There are no missing fields or unmeasured confounders within the experimental setup.
- Full researcher control. Centola controlled the network structure, the behavior being studied, the timing, and the conditions. This eliminates algorithmic confounding, there is no platform algorithm shaping the results behind the scenes.

Cons:
- Small scale. Custom-made experiments are expensive and time-consuming to set up, limiting sample size to around 1,500 participants. This makes it harder to detect small effects or study heterogeneity across subgroups.
- Weak external validity. The social network was artificial, and participants knew they were in an experiment. This *reactivity* (Salganik, Ch. 2.3.3) means their behavior may not reflect how people act in real social networks. It is unclear whether the findings generalize beyond the lab setting.
- Costly. Recruiting participants, building the platform, and running the experiment requires significant time and resources.

**Nicolaides's study (ready-made data):**

Pros:
- Big. The fitness app dataset contains millions of runners over years of activity, enabling the study of rare events, subgroup differences, and small effects (Salganik, Ch. 2.3.1).
- Always-on. The data was collected continuously and longitudinally, allowing researchers to track changes in running behavior over time without needing to plan data collection around specific events (Salganik, Ch. 2.3.2).
- Nonreactive. Runners were not aware their data would be used for research, so their behavior was not influenced by being observed (Salganik, Ch. 2.3.3).
- Strong external validity. The data reflects real people engaging in real behavior within genuine social networks, making the findings more likely to generalize to real-world settings.
- Low cost of data acquisition. The data already existed, no need to recruit participants or build infrastructure.

Cons:
- Weak internal validity. Without random assignment, the study can only show correlations, not causation. The observed "contagion" effect could be driven by confounders such as shared weather, local events, or similar lifestyles among connected individuals.
- Incomplete. The dataset likely lacks important variables such as runners' motivations, income, health conditions, or offline social interactions, information the app was never designed to capture (Salganik, Ch. 2.3.4).
- Algorithmically confounded. The fitness app may use notifications, challenges, leaderboards, or friend suggestions that artificially influence running behavior. These platform-driven effects are largely invisible to the researchers but could be mistaken for genuine social influence (Salganik, Ch. 2.3.8).
- Nonrepresentative. Users of a fitness tracking app are not a random sample of the general population. They are likely wealthier, healthier, younger, and more tech-savvy, which limits the generalizability of the findings (Salganik, Ch. 2.3.6).
- Dirty. The data may contain GPS errors, inactive accounts, or automated activity that does not reflect real running behavior (Salganik, Ch. 2.3.9).

### Q2

The key differences between the two data sources directly affect what conclusions can be drawn from each study.

Centola's custom-made experiment allows for strong causal claims: one can conclude that network structure causes differences in behavior spread, because all other variables were controlled through random assignment. However, because the experiment took place in an artificial setting with participants who knew they were being observed, we cannot be confident that these results hold in real-world social networks. The small scale also limits the ability to explore whether the effect varies across different types of people or contexts.

Nicolaides's study, by contrast, captures real-world behavior at massive scale, giving it strong ecological validity. But the lack of experimental control means the observed "exercise contagion" could have alternative explanations. For example, two friends might both increase their running not because one influenced the other, but because they live in the same city and experienced the same good weather, or because the app sent both of them a notification about a running challenge (algorithmic confounding). The incompleteness of the data, missing variables like motivation or health, makes it impossible to rule out such confounders. Furthermore, because fitness app users are not representative of the general population, the findings may not apply to people who do not use such apps.

In summary, Centola's study trades generalizability for causal certainty, while Nicolaides's study trades causal certainty for scale and realism. Neither approach alone gives a complete picture, which is why Salganik advocates for hybrid approaches that combine the strengths of custom-made and ready-made data.

## Exercise 2: APIs for Computational Social Science research

**Data description:** The Reddit Data API provides access to posts, comments, user profiles, subreddit metadata, and voting scores across Reddit's communities. This makes it useful for CSS research on public discourse, opinion formation, community dynamics, and sentiment analysis, for example, studying how privacy-related events shift discussion in subreddits like r/privacy or r/degoogle.

**Rate limits (free tier):** Free tier allows 100 requests per minute (QPM) for OAuth-authenticated applications, and 10 requests per minute for unauthenticated access. The QPM limit is averaged over a 10-minute window, so burst traffic is briefly allowed. The free tier is available for non-commercial uses such as personal projects and academic research. One important limitation: you can only paginate through about 1,000 posts per subreddit listing, which caps how much historical data you can pull.

### Helper func for single author

```python
def get_author_info(name):
    """Query OpenAlex API for an author name and return info for the top result."""
    url = "https://api.openalex.org/authors"
    params = {"search": name}
    response = requests.get(url, params=params)
    
    if response.status_code != 200:
        return None
    
    data = response.json()
    
    if not data['results']:
        return None
    
    top = data['results'][0]
    
    # extract country_code from last known institution (if available)
    institutions = top.get('last_known_institutions', [])
    country_code = institutions[0]['country_code'] if institutions else None
    
    return {
        'searched_name': name,
        'openalex_id': top['id'],
        'display_name': top['display_name'],
        'works_api_url': top['works_api_url'],
        'h_index': top['summary_stats']['h_index'],
        'works_count': top['works_count'],
        'country_code': country_code,
        'relevance_score': top['relevance_score']
    }
```

```python
# quick test
result = get_author_info("Duncan Watts")
result
```

### Loop through all names

```python
results = []
failed = []

for i, name in enumerate(names):
    try:
        info = get_author_info(name)
        if info:
            results.append(info)
        else:
            failed.append(name)
    except Exception as e:
        failed.append(name)
        print(f"Error for '{name}': {e}")
    
    # be polite to the API
    time.sleep(0.1)
    
    # progress update every 100 names
    if (i + 1) % 100 == 0:
        print(f"Processed {i + 1}/{len(names)} names ({len(results)} found, {len(failed)} failed)")

print(f"\nDone! Found: {len(results)}, Failed: {len(failed)}")
```

### Reflection Questions

**Which challenges did you encounter? How did you address them?**

We ran into a few issues. The main one was that the OpenAlex search endpoint returns up to 25 results per name, and the top result isn't always the right person. For instance, searching "Aaron Schein" gave us "Joshua Maynez", a completely different researcher. To handle this, we used fuzzy string matching to compare the searched name against the returned `display_name`, and dropped everything below 70% similarity. Another issue was rate limiting, after about 1,000 requests the API started returning 429 errors. Turns out the free tier only allows 1,000 search calls per day, so we registered for a free API key and retried the failed names. We also noticed some returned names were in non-Latin scripts (e.g., Cyrillic), giving misleadingly low similarity scores despite being correct. We accepted losing those as a trade-off. Lastly, 79 names had no results at all, likely due to scraping artifacts from Week 1 or researchers without an OpenAlex profile.

**Choose one problem you faced while collecting the data and describe your solution. Why did you choose this approach, and what impact might it have on your data?**

The biggest challenge was name disambiguation. The API often returned someone other than the person we searched for, sometimes a researcher with a similar name, sometimes someone completely unrelated. We first tried using the `display_name_alternatives` field, which contains name variants and transliterations, to improve matching. But we found that OpenAlex sometimes incorrectly merges different researchers into one record, so unrelated names show up in each other's alternatives. This made the approach unreliable, for example, "Aaron Schein" appeared in the alternatives of "Joshua Maynez". We went with a simpler solution instead: fuzzy matching between the searched name and `display_name`, dropping anything below 70%. This is conservative and means we lose some valid matches (especially Cyrillic names), but it avoids keeping clearly wrong entries in the dataset. We think a slightly smaller but more reliable dataset is better for any downstream analysis.
